Sea :
\begin{equation*}
\ddot{z} = -\dfrac{1}{z}\left[(\dot{z})^{2}+gz-gh+b\dot{z}\right]
\end{equation*}

la ecuacion diferencial que modela un liquido en un popote

definimos
\begin{equation*}
\left.\begin{cases} z_{1}=z\\z_{2}=\dot{z}\end{cases} \right.
\end{equation*}

\begin{equation*}
\dot{Z}(t)=\begin{pmatrix} \dot{z}\\ -\dfrac{1}{z}\left[(\dot{z})^{2}+gz-gh+b\dot{z}\right] \end{pmatrix}
\end{equation*}

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
import ipywidgets as ipw
```

```python
# para poder resolver esta ecuacion desacomplamos la misma
def ode(Z,t,b,g,h):
    z, dotz = Z
    ddotz = (dotz**2+g*z-g*h+b*dotz)*(-1/z)
    return [dotz, ddotz]

def update(b=0.1,g=9.81,h=0.102):
    Z0 = [0.02, 0.]
    sol = integrate.odeint(ode, Z0,t,args = (b,g,h))
    return sol[:, 0]

plt.plot(update(b=0.02,g=9.81,h=0.102))
plt.xlabel("Time",fontsize=10)
plt.ylabel("A",fontsize=10)
plt.show()
```
